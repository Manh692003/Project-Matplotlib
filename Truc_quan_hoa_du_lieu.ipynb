# Project: Trực quan hóa dữ liệu các thành phố ở California

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('california_cities.csv')
df.head()
```

```python
vi_do, kinh_do = df['latd'], df['longd']
dan_so, dien_tich = df['population_total'], df['area_total_km2']

plt.style.available
plt.style.use('seaborn-whitegrid')
plt.style.use('seaborn');
plt.figure(figsize = (10, 7))

plt.scatter(kinh_do, vi_do,
            c = np.log10(dan_so), cmap = 'viridis',
            s = dien_tich, linewidths = 0, alpha = 0.5);
plt.axis('equal');
plt.xlabel('Kinh độ')
plt.ylabel('Vĩ độ');
plt.colorbar(label = 'Log$_{10}$(Dân số)')
plt.clim(3, 7);

chu_thich = [50, 100, 300, 500]
for dt in chu_thich:
    plt.scatter([], [], s = dt, label = str(dt) + ' km$^2$', c = 'k', alpha = 0.4)
    
plt.legend(labelspacing = 1, title = 'Diện tích');

plt.title('Trức quan hóa dữ liệu các thành phố ở Calofornia');
```
